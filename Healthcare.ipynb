##### There are 768 observations and 9 features/variables. Minimum values for variable 'Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI' are 0 which doesn't make sense and thus indicates missing value.

```python
missing_col = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
for col in missing_col:
    data[col] = data[col].replace(0, np.nan)
```

```python
data.isnull().sum()
```

```python
data.to_excel('healthcare.xlsx', sheet_name = 'healthcare', index = False)
```

```python
for col in missing_col:
    print(col)
    print('Mean is: ', data[col].mean())
    print('Median is: ', data[col].median())
    print('Mode is: ', data[col].mode())
    plt.figure(figsize = (8, 5))
    sns.distplot(data[col], color = 'green')
    plt.show()
```

```python
data['Pregnancies'].value_counts(dropna = False)
```

##### Based on histogram distribution, for feature 'Pregnancies', 1 is the most frequency, so I will use mode strategy to treat missing value. I will use median strategy for  feature 'Insulin' , and mean strategy for the others features.

```python
from sklearn.impute import SimpleImputer
impute_mean = SimpleImputer(missing_values = np.nan, strategy = 'mean')
impute_median = SimpleImputer(missing_values = np.nan, strategy = 'median')
impute_mode = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
```

```python
data[['Pregnancies']] = impute_mode.fit_transform(data[['Pregnancies']])
data[['Insulin']] = impute_median.fit_transform(data[['Insulin']])
data[['Glucose', 'BloodPressure', 'SkinThickness', 'BMI']] = impute_mean.fit_transform(data[['Glucose', 'BloodPressure', 'SkinThickness', 'BMI']])
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
```

```python
X = data.drop('Outcome', axis = 1)
y = data[['Outcome']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 2)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
lg = LogisticRegression()
lg.fit(X_train_scaled, y_train)
y_pred = lg.predict(X_test_scaled)
prob = lg.predict_proba(X_test_scaled)
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
print(classification_report(y_test, y_pred))
```

```python
# Use score method to get accuracy of model
score = lg.score(X_test_scaled, y_test)
print(score)
```

```python
plt.figure(figsize = (8, 8))
sns.heatmap(cm, annot=True, fmt=".2f", linewidths=.5, cmap = 'PiYG')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.title('Accuracy Score: {}'.format(round(score*100, 2)))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
```

```python
models = ['Nearest Neighbors', 'Decision Tree', 'Naive Bayes', 'Support Vector Machine', 
          'Random Forest', 'AdaBoost', 'Gradient Boost', 'XGBoost']
classifiers = [KNeighborsClassifier(),
               DecisionTreeClassifier(max_depth = 5),
               GaussianNB(),
               SVC(probability = True, ),
               RandomForestClassifier(max_depth = 5, n_estimators = 100),
               AdaBoostClassifier(),
               GradientBoostingClassifier(),
               XGBClassifier()]
```

```python
for model, classifier in zip(models, classifiers):
    print(model)
    classifier.fit(X_train_scaled, y_train)
    y_pred = classifier.predict(X_test_scaled)
    prob = classifier.predict_proba(X_test_scaled)
    score = classifier.score(X_test_scaled, y_test)
    print('Accuracy Score: ', round(score*100, 2))
    print()
    
    
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize = (8, 8))
    sns.heatmap(cm, annot = True, fmt = '.2f', linewidth = .5, cmap = 'rainbow')
    plt.ylabel('Actual label')
    plt.xlabel('Predicted label')
    plt.title('Accuracy Score: {}'.format(round(score*100, 2)))
    print(classification_report(y_test, y_pred))
    print()
    
    skplt.metrics.plot_roc_curve(y_test, prob, figsize = (8, 6), title = 'ROC Curve for: {}'.format(model))
    plt.show()
    print('********************************************************************')
    print()

    
```
